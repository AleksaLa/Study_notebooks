Отдельных кухонь меньше пяти квадратов не бывает, поэтому в несколько-комнатых квартирах - это объединенная кухня с гостиной, тн евродвушки и евротрешки, а однокомнатные - это студии. Поменяем их тип булевого значения на "студия правда".

```python
# Поменяем тип булевого значения студий
data['studio'] = data['studio'].where(~((data['kitchen_area'] < 5) & (data['rooms'] == 1)), True)
```
